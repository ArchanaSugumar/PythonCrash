### Tensors-FF1

Create a tensor of size (3, 4) filled with random numbers and print it.

```python
import torch

# Create a tensor of size (3, 4) filled with random numbers
tensor = torch.rand(3, 4)
print(tensor)
```

### Tensors-FF2

Reshape a tensor of shape (2, 3, 4) to a tensor of shape (6, 4) and print the new shape.

```python
import torch

# Create a tensor of shape (2, 3, 4)
tensor = torch.rand(2, 3, 4)

# Reshape the tensor to shape (6, 4)
reshaped_tensor = tensor.reshape(6, 4)
print(reshaped_tensor.shape)
```

### Tensors-FF3

Perform element-wise addition of two tensors of the same shape (4, 5).

```python
import torch

# Create two tensors of shape (4, 5)
tensor1 = torch.rand(4, 5)
tensor2 = torch.rand(4, 5)

# Perform element-wise addition
result = tensor1 + tensor2
print(result)
```
